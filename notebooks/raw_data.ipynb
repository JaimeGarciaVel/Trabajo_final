```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
data_pca = df_datos_socioeconomicos[[x for x in df_datos_socioeconomicos.columns if x != 'Perfil_inversor']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_rescaled = scaler.fit_transform(data_pca)

data_rescaled
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42).fit(data_rescaled.data)

kmeans
```

```python
clusters=kmeans.predict(data_rescaled.data)
clusters
```

```python
kmeans.cluster_centers_[0]
```

```python
import numpy as np

plt.scatter(data_rescaled[:, 0], data_rescaled[:, 1], s=5)
plt.gca().set_aspect('equal', 'datalim')
plt.title('Embedding projection', fontsize=18);
```

```python
plt.scatter(data_rescaled[:, 0], data_rescaled[:, 1], c=kmeans.labels_, cmap='Spectral', s=5)
plt.gca().set_aspect('equal', 'datalim')
plt.colorbar(boundaries=np.arange(kmeans.n_clusters + 1) - 0.5).set_ticks(np.arange(kmeans.n_clusters))
plt.title('K-Means over the datos_socioeconomicos dataset', fontsize=20);
```

```python
from sklearn.cluster import DBSCAN

dbscan_c = DBSCAN(eps=1.5,
                  min_samples=25).fit(data_rescaled)

dbscan_c
```

```python
plt.scatter(data_rescaled[:, 0], data_rescaled[:, 1], c=dbscan_c.labels_, cmap='Spectral', s=5)
plt.gca().set_aspect('equal', 'datalim')
plt.colorbar(boundaries=np.arange(len(np.unique(dbscan_c.labels_)) + 1) - 0.5).set_ticks(np.arange(len(np.unique(dbscan_c.labels_))))
plt.title('DBSCAN over the datos_socioeconomicos dataset', fontsize=20);
```
